```python
def f(x) :
    if x==0 : return   #종료 조건
    f(x-1)
    print(x)
    
n = int(input())
f(n)
```

# 1~n까지 합

```python
def f(x) :
    if x == 1 : return 1  #종료 조건
    return f(x-1) + x

n = int(input())
res = f(n)
print(res)
```
